```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("/content/AI_Student_Life_Pakistan_2026.csv")

# Preview data
df.head()
```

```python
df.info()
df.describe()
df.columns
```

```python
# Check missing values
df.isnull().sum()

# Remove duplicates
df = df.drop_duplicates()
```

```python
plt.hist(df['Daily_Usage_Hours'], bins=10)
plt.title("Daily Usage of AI/Social Media")
plt.xlabel("Hours")
plt.ylabel("Number of Students")
plt.show()
```

```python
df['AI_Tool_Used'].value_counts().plot(kind='bar')
plt.title("Most Used AI Tools / Platforms")
plt.show()
```

```python
sns.boxplot(x='Gender', y='Daily_Usage_Hours', data=df)
plt.title("Usage by Gender")
plt.show()
```

```python
df['Impact_on_Grades'].value_counts().plot(kind='bar')
plt.title("Academic Impact")
plt.show()
```

```python
sns.boxplot(x='Impact_on_Grades', y='Daily_Usage_Hours', data=df)
plt.title("Usage vs Academic Impact")
plt.show()
```

```python
sns.scatterplot(x='Daily_Usage_Hours', y='Satisfaction_Level', data=df)
plt.title("Usage vs Mental Health")
plt.show()
```

Key Findings

	•	Students with higher usage tend to have lower sleep hours
	•	Increased usage is linked with higher stress levels
	•	Moderate usage shows balanced outcomes
	•	Excessive use negatively impacts academic performance
